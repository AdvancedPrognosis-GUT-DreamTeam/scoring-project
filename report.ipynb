```python
import pandas as pd

data = pd.read_excel("data/kredyty_auto_Scoring2025s.xlsx")
data.head(10)
```

```python
def calc_annuit_rate(row):
    credit = row['kwota_kredytu']  # kwota kredytu
    yearly_percentage = row['oproc_propon']  # oprocentowanie roczne w procentach
    n = row['okres_kredytu']  # liczba miesięcy

    r = (yearly_percentage / 100) / 12  # miesięczna stopa procentowa jako ułamek

    # wzór na ratę annuitetową:
    rata = credit * (r * (1 + r)**n) / ((1 + r)**n - 1)
    return rata
data['rata_annuitowa'] = data.apply(calc_annuit_rate, axis = 1)
data.head()
```
